```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Sales.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = 'mixed', dayfirst = True)
```

```python
df.info()
```

```python
reference_date = df['Date'].max()
reference_date
```

```python
recency_df = df.groupby('Customer Name')['Date'].max().reset_index()
recency_df['recency'] = (reference_date - recency_df['Date']).dt.days 
recency_df.head()
```

```python
frequency_df = df.groupby('Customer Name').size().reset_index(name = 'Frequency')
frequency_df
```

```python
monetary_df = df.groupby('Customer Name')['Total Sales'].sum().reset_index(name= 'Monetary')
monetary_df
```

```python
rf_df = recency_df.merge(frequency_df, on ='Customer Name')
rf_df
```

```python
rfm = rf_df.merge(monetary_df, on = 'Customer Name')
rfm
```

```python
def segment_customer(row):
    if row['recency'] <= 30 and row['Frequency'] >= 5 and row['Monetary'] >= 50000:
        return 'Champion'
    elif row['Frequency'] >= 3 and row['Monetary'] >= 20000:
        return 'Loyal'
    else:
        return 'Hibernating'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
```

```python
rfm
```

```python
rfm.to_csv('rfm_customer360.csv',index=False)
```
